# Manipulation de données CSV avec pandas

```python
from os import listdir
assert 'train.csv' in listdir("."), "Il manque le jeu de données exemple."
```

## Lire un fichier CSV

```python
import pandas as pd

df = pd.read_csv("./train.csv")
print("Jeux de données chargé ; " + str(df.shape[0]) + " lignes, " + str(df.shape[1]) + " colonnes.")
```

## Afficher un sous ensemble des données

```python
from IPython.display import HTML
vingts_premiers = df[0:20] # on extrait les 20 premières lignes
HTML(vingts_premiers.to_html())
```

## Sélectionner une colonne

```python
print(df)
```

```python
age_df = df["Age"]
print(age_df[0:20])
# Deux manières équivalentes d'adresser les colonnes
assert df.Age is age_df
```

```python
# Utilisation des index
dix_premiers_classe_nom_sexe = df.iloc[:10, 2:5]

dix_nom_ticket = df.loc[10:20,"Name":"Ticket"]
HTML(dix_nom_ticket.to_html())
```

# Simple filtre

```python
plus_de_20 = df[ df.Age > 20 ]
HTML(plus_de_20[0:10].to_html())
```
